```python
from nba_api.stats.endpoints import leaguegamefinder
```

```python
gamefinder = leaguegamefinder.LeagueGameFinder(date_to_nullable='06/01/2023', league_id_nullable='00')
games = gamefinder.get_data_frames()[0]
games.head()
```

```python
games.columns
```

```python
games = games[['TEAM_NAME', 'GAME_ID','GAME_DATE', 'MATCHUP', 'WL','PLUS_MINUS']]
```

```python
games
```

```python
games = games.sort_values('GAME_DATE')
```

```python
games
```

```python
games['avg_30_plus_minus']=games.groupby('TEAM_NAME')['PLUS_MINUS'].transform(lambda x: x.rolling(30, closed='left').mean())
```

```python
games[games['TEAM_NAME']=='Miami Heat'].head(35)
```

```python
msk = games['MATCHUP'].str.contains('@')
```

```python
games_away = games[msk]
games_home = games[~msk]
```

```python
games_home
```

```python
games_away
```

```python
games_home.shape
```

```python
games_away.shape
```

```python
import pandas as pd
games_merged = pd.merge(games_home,games_away,on='GAME_ID',suffixes=('_home','_away'))
```

```python
games_merged
```

```python
games_merged['avg_30_plus_minus_diff']=games_merged['avg_30_plus_minus_home']-games_merged['avg_30_plus_minus_away']
```

```python
games_merged
```

```python
games_model = games_merged[['WL_home','avg_30_plus_minus_diff']].dropna()
games_model['WL_home'] = games_model['WL_home'].map({'W':1,'L':0})
```

```python
games_model
```

```python
from sklearn.model_selection import train_test_split
```

```python
df_train,df_test=train_test_split(games_model,stratify=games_model['WL_home'],test_size=0.2,random_state=7)
```

```python
df_train.shape
```

```python
df_test.shape
```

```python
target = 'WL_home'
X_train = df_train.drop(columns = target)
y_train = df_train[target]

X_test = df_test.drop(columns = target)
y_test = df_test[target]
```

```python
import xgboost as xgb
```

```python
clf = xgb.XGBClassifier(random_state = 7)
clf.fit(X_train,y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
#hyperparameter tuning
from sklearn.model_selection import RandomizedSearchCV
```

```python
params = {'learning_rate': (0.0001,1),
         'n_estimators': [50,100,150,200,250],
         'max_depth': [2,3,4,5,6,7,8,9],
         'subsample': [0.7,0.8,0.9,1.0]}
```

```python
Random_P = RandomizedSearchCV(estimator = clf, param_distributions = params, n_iter = 20, cv = 7, scoring = 'accuracy', random_state = 7)
```

```python
Random_P.fit(X_train,y_train)
```

```python
Random_P.best_params_
```

```python
Model_p = Random_P.best_estimator_
y_pred_p = Model_p.predict(X_test)
```

```python
accuracy_score(y_test,y_pred_p)
```
